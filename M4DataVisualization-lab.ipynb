# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
query = """
SELECT ConvertedComp FROM master
"""
hist = pd.read_sql_query(query,conn)
hist.plot(kind ='hist',bins=20)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

query = """
SELECT Age FROM master
"""
age = pd.read_sql_query(query,conn)
sns.boxplot(x=age,data=age)
plt.xlabel('Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query = """
SELECT Age,WorkWeekHrs FROM master
"""

sc = pd.read_sql_query(query,conn)

sns.scatterplot(x='Age',y='WorkWeekHrs', data= sc)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
query = """
SELECT Age,WorkWeekHrs,CodeRevHrs FROM master
"""

bubble = pd.read_sql_query(query,conn)
norm_age= (bubble['Age'] - bubble['Age'].min()) / (bubble['Age'].max() - bubble['Age'].min())

bubble.plot(kind='scatter',
                    x='WorkWeekHrs',
                    y='CodeRevHrs',
                    figsize=(14, 8),
                    alpha=0.3,  # transparency
                    s=norm_age * 2000 + 10,  # pass in weights 
                    )
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query = """
SELECT count(*)as count, DatabaseDesireNextYear 
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear 
order by count DESC limit 5
"""

top5_databases= pd.read_sql_query(query,conn)


label_list =top5_databases['DatabaseDesireNextYear']

top5_databases['count'].plot(kind= 'pie',
                                  figsize=(10,6),
                                  startangle=90,
                                  autopct='%1.1f%%',
                                  labels = label_list,
                                  title= 'Top 5 Desired Databses'
                                 )
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

query = """
SELECT Age,CodeRevHrs,WorkWeekHrs FROM master
WHERE Age>=30 and Age<=35

"""

stacked_df = pd.read_sql_query(query,conn)
stacked_df = stacked_df.groupby('Age')[['CodeRevHrs','WorkWeekHrs']].median().reset_index()
stacked_df['Age']=stacked_df['Age'].astype(int)
stacked_df.plot(kind='bar',
                x='Age',
                figsize=(10,6)
               )
plt.ylabel('Working hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

query = """
SELECT Age,ConvertedComp FROM master
WHERE Age>=45 and Age<=60

"""

line1= pd.read_sql_query(query,conn)
line1= line1.groupby('Age')['ConvertedComp'].median()
line1.plot(kind='line',
           x='Age',
           y='ConvertedComp',
           figsize=(10,6)
          )
plt.grid()
plt.ylabel('Converted Compensation')
plt.title('Median Converted Compensation for the Age group of 45-60')
plt.show()
```
